```python
import pickle

file_paths = ['MLP_roc.pkl','CNN_roc.pkl','RBFN_roc.pkl']

loaded_data = []

#load data from each file
for file_path in file_paths:
    with open(file_path,'rb') as f:
        data = pickle.load(f)
    loaded_data.append(data)   
```

```python
import matplotlib.pyplot as plt
plt.plot(
    loaded_data[0]['MLP_fpr_macro'],
    loaded_data[0]['MLP_tpr_macro'],
    label=f"MLP (AUC = {loaded_data[0]['MLP_auc_macro']:.3f})",
    color="cornflowerblue",
    linewidth=2,
)


plt.plot(
    loaded_data[1]['CNN_fpr_macro'],
    loaded_data[1]['CNN_tpr_macro'],
    label=f"CNN (AUC = {loaded_data[1]['CNN_auc_macro']:.3f})",
    color="darkorange",
    linewidth=2,
)

plt.plot(
    loaded_data[2]['RBFN_fpr_macro'],
    loaded_data[2]['RBFN_tpr_macro'],
    label=f"RBFN (AUC = {loaded_data[2]['RBFN_auc_macro']:.3f})",
    color="purple",
    linewidth=2,
)


plt.legend(loc="lower right")
plt.axis("square")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Multiclass AUROC for different Classifiers in Trait Classification")
```
